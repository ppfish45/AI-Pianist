```python
fix, ax = plt.subplots(1, 16, figsize=(10, 30))
for inputs, labels in dataset_p.data_batch(type='val', size='single', color='black', batch_size=16, NCHW=False):
    for i in range(16):
        ax[i].axis('off')
        ax[i].imshow(cv2.cvtColor(inputs[i], cv2.COLOR_BGR2RGB))
        ax[i].set_title(labels[i])
    break
```

```python
fix, ax = plt.subplots(1, 18, figsize=(10, 30))
for inputs, labels in dataset_p.data_batch(type='val', size='single', color='black', batch_size=6, NCHW=False,
                                          concatenate=True, shuffle=True):
    for i in range(6):
        ax[i * 3].axis('off')
        ax[i * 3 + 1].axis('off')
        ax[i * 3 + 2].axis('off')
        ax[i * 3].imshow(cv2.cvtColor(inputs[i, :, :, 0:3], cv2.COLOR_BGR2RGB))
        ax[i * 3 + 1].imshow(cv2.cvtColor(inputs[i, :, :, 3:6], cv2.COLOR_BGR2RGB))
        ax[i * 3 + 2].imshow(cv2.cvtColor(inputs[i, :, :, 6:9], cv2.COLOR_BGR2RGB))
        ax[i * 3 + 1].set_title(labels[i])
    break
```
