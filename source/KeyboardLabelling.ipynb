```python
import cv2
import matplotlib.pyplot as plt
import skimage
import numpy as np
import imageio
import copy

import my_utils
import dataset
```

```python
v = dataset.video(1, True)
n_frame = v.get_num_of_frames()
sample = v.get_frame(30)
plt.figure(figsize=(12, 24))
plt.imshow(sample)
```

```python
'''
(x1, y1) - top left corner
(x2, y2) - top right corner
(x3, y3) - bottom left corner
(x4, y4) - bottom right corner
'''
def label(src_image, x, y):
    image = copy.copy(src_image)

    for i in range(4):
        cv2.circle(image, (x[i], y[i]), 2, (250, 50, 2), 2)
    
    plt.figure(figsize=(12, 24))
    plt.imshow(image)
```

```python
label(sample, [630, 4, 630, 4], [205, 202, 128, 130])
```
