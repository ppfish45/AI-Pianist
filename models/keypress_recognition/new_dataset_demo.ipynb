```python
import cv2
import dataset_p
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset_p.load_all_data(spliter=['val'], color=['black', 'white'], size=['single', 'bundle'], keypress=True, shuffle=True)
```

```python
for inputs, labels in dataset_p.data_batch(type='val', size='single', color='white', batch_size=8, NCHW=False):
    for i in range(8):
        plt.imshow(cv2.cvtColor(inputs[i], cv2.COLOR_BGR2RGB))
        plt.title(labels[i])
        plt.show()
    break
```

```python
for inputs, labels in dataset_p.data_batch(type='val', size='bundle', color='white', batch_size=8, need_velocity=False, NCHW=False):
    for i in range(8):
        plt.imshow(cv2.cvtColor(inputs[i], cv2.COLOR_BGR2RGB))
        plt.title(labels[i])
        plt.show()
    break
```
