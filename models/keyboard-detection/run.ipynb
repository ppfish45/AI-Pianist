```python
import copy
import model
import dataset
import numpy as np
import matplotlib.pyplot as plt
```

```python
# print several keyboard images with labels

for _type in ['train', 'test', 'val']:

    print(f'Samples from X_{_type}: ')

    fig, axarr = plt.subplots(1, 2, figsize=[16, 16])

    for i in range(0, 2):
        img, pts = dataset.get_sample(_type)
        img = dataset.show_labelled_image(img, pts)
        axarr[i].imshow(img)

    plt.show()
```

```python
# print several RESIZED keyboard images with labels

for images, labels in dataset.data_batch(batch_size=4, image_size=(224, 224), NCHW=False):
    
    fig, axarr = plt.subplots(1, 2, figsize=[16, 16])

    for i in range(0, 2):
        img = images[i]
        pts = labels[i]
        img = dataset.show_labelled_image(img, pts)
        axarr[i].imshow(img)

    plt.show()
    
    break
```

```python
# print several RESIZED and ROTATED keyboard images with labels

for images, labels in dataset.data_batch(batch_size=4, image_size=(224, 224), dirs=[0, 2], NCHW=False):
    
    fig, axarr = plt.subplots(1, 2, figsize=[16, 16])

    for i in range(0, 2):
        img = images[i]
        pts = labels[i]
        img = dataset.show_labelled_image(img, pts)
        axarr[i].imshow(img)

    plt.show()
    
    break
```
