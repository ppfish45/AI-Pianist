```python
from keypress_recognition import separate as sp
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
img = cv2.imread("../keyboard_3.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
print(img.shape)
plt.imshow(img)
white_bounding1, black_bounding1 = sp.get_bounding_box(img)
white_bounding2, black_bounding2 = sp.get_bounding_box2(img)
```

```python
imm1 = img.copy()
imm2 = img.copy()
color_switch = False
print(black_bounding2)
print(len(black_bounding2))
for left, right, up, down in black_bounding1:
    color = (0, 255, 0) if color_switch else (0, 0, 255)
    color_switch = not color_switch
    cv2.line(imm1, (left, up), (left, down), color, 1)
    cv2.line(imm1, (right, up), (right, down), color, 1)
for left, right, up, down in black_bounding2:
    color = (0, 255, 0) if color_switch else (0, 0, 255)
    color_switch = not color_switch
    cv2.line(imm2, (left, up), (left, down), color, 1)
    cv2.line(imm2, (right, up), (right, down), color, 1)
fig, ax = plt.subplots(2, 1, figsize=(20, 10))
ax[0].imshow(imm1)
ax[1].imshow(imm2)
```

```python
for xx in white_bounding2:
    assert xx[1] - xx[0] == 21
for xx in black_bounding2:
    assert xx[1] - xx[0] == 16
```
