```python
from keypress_recognition.seperate import get_bounding_box
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
img = cv2.imread("keyboard_example.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
print(img.shape)
plt.imshow(img)
white_bounding, black_bounding = get_bounding_box(img)
```

```python
# 
plt.imshow(cv2.rectangle(img.copy(), (1, 1), (50, 200), (255, 0, 0), 1))
f, ax = plt.subplots(5, 2)
for i, (left, right, up, down) in enumerate(white_bounding[:10]):
    irow = i // 2
    icol = i % 2
    print(left, right, up, down)
    img_with_box = cv2.rectangle(img.copy(), (left, up), (right, down), (255, 0, 0), 1)
    ax[irow, icol].imshow(img_with_box)
```
