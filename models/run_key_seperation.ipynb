```python
from keypress_recognition import separate as sp
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
img = cv2.imread("../keyboard.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
print(img.shape)
plt.imshow(img)
white_bounding, black_bounding = sp.get_bounding_box(img)
```

```python
# 
imm = img.copy()
color_switch = False
print(len(black_bounding))
for left, right, up, down in black_bounding:
    color = (0, 255, 0) if color_switch else (0, 0, 255)
    color_switch = not color_switch
    cv2.line(imm, (left, up), (left, down), color, 1)
    cv2.line(imm, (right, up), (right, down), color, 1)
fig, ax = plt.subplots(figsize=(20, 10))
ax.imshow(imm)
```
