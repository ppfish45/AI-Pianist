```python
for arr, label in dataset.lstm_data_batch(type='val', color='white'):
    print(label)
    arr = np.transpose(arr, (0, 2, 3, 1))
    fix, ax = plt.subplots(1, len(arr) + 2)
    print(arr.shape)
    for i, img in enumerate(arr):
        ax[i + (i >= 4) + (i >= len(arr) - 4)].axis('off')
        if i > 0:
            ax[i + (i >= 4) + (i >= len(arr) - 4)].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        else:
            ax[i + (i >= 4) + (i >= len(arr) - 4)].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    break
```
